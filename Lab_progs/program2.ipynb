```python
import pandas as pd
import numpy as np
```

```python
#import the dataset
path = "./datasets/P2_mov.csv"
df = pd.read_csv(path)
```

```python
df.head()
```

```python
# data inspection
print("list of columns: ",df.columns)
print("number of rows: ",len(df))
```

```python
# Data Reduction: choosing required columns:
colms = ['budget','genres','original_language','title','overview','release_date','runtime','status','vote_average']
df = df[colms]
```

```python
df.head()
```

```python
df.describe()
```

```python
# run time missing values
df[pd.isnull(df["runtime"])]
```

```python
# filled missing value with mean
df["runtime"] = df["runtime"].fillna(df["runtime"].mean())
```

```python
df.head()
```

```python
# no more missing values
df[pd.isnull(df["runtime"])]
```

```python
#data descritization
df['status'] = np.where(df['vote_average']>=6,'HIT','FLOP')
```

```python
df.head()
```

```python
#normalizing budget (absolute maximum scaling)
df["revenue_scaled"] = df['budget']/df['budget'].abs().max()
```
